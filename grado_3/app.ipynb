```python
import os

lessons_folder = "./lecciones"
lessons = os.listdir(lessons_folder)
lessons
```

```python
import numpy as np

lesson_name_sections_counts = []
lesson_name_sections_examples = []

for lesson in lessons:
    lesson_name_sections = lesson.split("-")
    
    lesson_sections_number = len(lesson_name_sections)
    
    if lesson_sections_number not in lesson_name_sections_counts:
        lesson_name_sections_examples.append(lesson_name_sections)

    lesson_name_sections_counts.append(lesson_sections_number)
    
    lesson_sections_number_ = str(len(lesson_name_sections)).zfill(2)
    print(f"Name sections: {lesson_sections_number} ===> {lesson_name_sections}")

    
```

```python
lesson_name_sections_counts = np.unique(lesson_name_sections_counts)
print(f"\n{len(lesson_name_sections_counts)} file names' patterns found")
print(f"\nNumber of file names' sections:\n{'-'*66}")
for n in lesson_name_sections_counts:
    for example in lesson_name_sections_examples:
        if len(example) == n:
            print(f"Sections: {n:>2} {example}")
```

```python
import re
import os

degree = "03"

# Improved regex pattern to match all numbers after "leccion" or "lección"
pattern = re.compile(r'(lección|leccion)(?:[-\s]*no\.)?[-\s]*(\d+)(?:[-\s\w]*?(\d+))?.*\.pdf', re.IGNORECASE)

# Function to rename files based on the pattern
def rename_file(file_name):
    match = pattern.search(file_name)
    if match:
        # Capture the first and second numbers
        first_number = match.group(2)
        last_number = match.group(3) if match.group(3) else first_number
        
        # Conditional logic: If there are two numbers, use the first one
        # This handles cases like '12-1' where we want to prioritize '12'
        if match.group(3):
            if int(last_number) < int(first_number):
                last_number = first_number
        
        # Create the new file name using the last significant number
        new_name = f"d{degree}_{int(last_number)}.pdf"  # Convert number to int to remove leading zeros
        return new_name
    return None

# Process and rename each file
for file_name in lessons:
    new_name = rename_file(file_name)
    if new_name:
        print(f"Renaming '{file_name}' to '{new_name}'")
        # Uncomment the following line to actually rename the files on disk
        # os.rename(f"{lessons_folder}/{file_name}", f"{lessons_folder}/{new_name}")
```

```python
# Extracting not found individual lessons from the book pdf containig all lessons:
# Not found:
#   lesson 11

from PyPDF2 import PdfReader, PdfWriter

def extract_pages(input_pdf, output_pdf, start, end):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    for page in range(start, end + 1):
        writer.add_page(reader.pages[page])

    with open(output_pdf, 'wb') as output_file:
        writer.write(output_file)

# Usage
extract_pages("./libro/d03_book.pdf", "./lecciones/d03_11.pdf", 113, 120)  # Extract pages 114 to 121
```

```python
def txt_to_html(filepath):
    html_content = ""

    with open(filepath, 'r', encoding='utf-8') as file:
        text = file.read()
        
        # Split the text into paragraphs (assuming paragraphs are separated by two newlines)
        paragraphs = text.split('\n\n')
        
        # Wrap each paragraph in <p> tags
        for paragraph in paragraphs:
            # Replace single newlines within paragraphs with <br> tags
            html_paragraph = paragraph.replace('\n', '<br>\n')
            html_content += f"<p>{html_paragraph}</p>\n"
    
    return html_content

# Usage
filepath = './test.txt'
html_output = txt_to_html(filepath)
print(html_output)  # You can save this to an HTML file if needed.
```

PDF to TEXT METHOD 1

```python
import PyPDF2

# Open the PDF file
with open('./lecciones/d03_1.pdf', 'rb') as pdf_file:
    reader = PyPDF2.PdfReader(pdf_file)
    
    # Initialize an empty string for the text
    extracted_text = ""
    
    # Loop through all the pages and extract text
    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        extracted_text += page.extract_text()

print(extracted_text)
```

```python
# Function to convert text to HTML format
def text_to_html(input_file, output_file):
    # Open and read the text file
    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Split the content into paragraphs (splitting by double newlines)
    paragraphs = content.split('\n\n')
    
    # Open the output HTML file for writing
    with open(output_file, 'w', encoding='utf-8') as f:
        # Write the basic structure of the HTML file
        f.write('<html>\n<head>\n<title>Converted Text</title>\n</head>\n<body>\n')
        
        # Convert each paragraph into HTML <p> tags
        for paragraph in paragraphs:
            # Strip leading/trailing whitespace and add <p> tags
            cleaned_paragraph = paragraph.strip()
            if cleaned_paragraph:
                f.write(f'<p>{cleaned_paragraph}</p>\n')
        
        # Close the HTML structure
        f.write('</body>\n</html>')

# Call the function with your input text file and output HTML file
input_txt = './contenido_limpio/d03_1.txt'  # Replace with your text file path
output_html = './contenido_limpio_html/d03_1.html'  # Replace with your desired output HTML file path
text_to_html(input_txt, output_html)
```
